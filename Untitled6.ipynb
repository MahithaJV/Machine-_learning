```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, f1_score, roc_auc_score, roc_curve)
from xgboost import XGBClassifier

print("Imports done ✅")


# ── CELL 3: Load data ─────────────────────────────────────────
df = pd.read_csv("hek_flares.csv")
df["event_starttime"] = pd.to_datetime(df["event_starttime"])
df["event_peaktime"]  = pd.to_datetime(df["event_peaktime"])
df["event_endtime"]   = pd.to_datetime(df["event_endtime"])
df = df.sort_values("event_starttime").reset_index(drop=True)
print("Loaded:", df.shape)


# ── CELL 4: SoLEXS 2.8 keV constraint ─────────────────────────
SOLEXS_FLUX_THRESHOLD = 1e-7
df = df[df["flux_estimate_wm2"] >= SOLEXS_FLUX_THRESHOLD].copy()
print(f"After SoLEXS filter: {df.shape[0]} events")


# ── CELL 5: Build a regular time grid ─────────────────────────
# Forecasting needs a continuous timeline, not just "flare rows".
# We bin time into 10-minute steps across the full date range,
# and at each step ask: "did/will a flare happen near here?"

df["log_flux"] = np.log10(df["flux_estimate_wm2"])
df["flare_class"] = df["fl_goescls"].str[0]
df["is_significant"] = df["flare_class"].isin(["C", "M", "X"]).astype(int)

start = df["event_starttime"].min().floor("10min")
end   = df["event_starttime"].max().ceil("10min")
timeline = pd.date_range(start, end, freq="10min")
grid = pd.DataFrame({"time": timeline})
print(f"Timeline: {len(grid)} steps of 10 min, "
      f"from {start} to {end}")


# ── CELL 6: Past-only rolling features at every timeline step ─
# For each timestamp t, we look BACKWARD only (no future leakage):
#  - time since last flare
#  - flux of last 1/2/3 flares
#  - rolling mean/std of last 3 flares
#  - count of flares in past 6h / 24h
#  - hour of day, day of year

flare_times = df["event_starttime"].values
flare_log_flux = df["log_flux"].values
flare_sig = df["is_significant"].values

def build_features(t_array):
    feats = []
    fi = 0  # pointer into flare events (flares are sorted)
    n = len(flare_times)
    # use searchsorted for speed
    idxs = np.searchsorted(flare_times, t_array, side="right") - 1  # last flare index <= t
    for i, t in enumerate(t_array):
        j = idxs[i]  # index of most recent flare at/before time t
        if j < 0:
            feats.append([np.nan]*11)
            continue
        last_time = flare_times[j]
        time_since_last = (t - last_time) / np.timedelta64(1, "m")

        prev1 = flare_log_flux[j]
        prev2 = flare_log_flux[j-1] if j-1 >= 0 else np.nan
        prev3 = flare_log_flux[j-2] if j-2 >= 0 else np.nan

        window = flare_log_flux[max(0, j-2): j+1]
        roll_mean = np.nanmean(window) if len(window) else np.nan
        roll_std  = np.nanstd(window) if len(window) > 1 else 0

        # counts in past 6h and 24h
        t64 = np.datetime64(t)
        mask_6h  = (flare_times <= t64) & (flare_times > t64 - np.timedelta64(6, "h"))
        mask_24h = (flare_times <= t64) & (flare_times > t64 - np.timedelta64(24, "h"))
        count_6h  = mask_6h.sum()
        count_24h = mask_24h.sum()
        sig_count_24h = flare_sig[mask_24h].sum()

        hour = pd.Timestamp(t).hour
        doy  = pd.Timestamp(t).dayofyear

        feats.append([time_since_last, prev1, prev2, prev3,
                      roll_mean, roll_std, count_6h, count_24h,
                      sig_count_24h, hour, doy])
    return np.array(feats)

print("Building features for each timeline step (this may take ~1-2 min)...")
X_all = build_features(grid["time"].values)

feature_cols = ["time_since_last_min", "prev_flux_1", "prev_flux_2", "prev_flux_3",
                 "rolling_mean_3", "rolling_std_3", "count_flares_6h",
                 "count_flares_24h", "count_significant_24h", "hour", "day_of_year"]
feat_df = pd.DataFrame(X_all, columns=feature_cols)
grid = pd.concat([grid, feat_df], axis=1)
grid.dropna(inplace=True)
grid.reset_index(drop=True, inplace=True)
print(f"Feature grid ready: {grid.shape}")


# ── CELL 7: Build FUTURE labels (the actual targets) ──────────
# Label = 1 if a C/M/X flare starts within the next N minutes from time t

def label_future_window(t_array, window_minutes):
    labels = np.zeros(len(t_array), dtype=int)
    sig_times = flare_times[flare_sig == 1]
    for i, t in enumerate(t_array):
        t64 = np.datetime64(t)
        future_end = t64 + np.timedelta64(window_minutes, "m")
        mask = (sig_times > t64) & (sig_times <= future_end)
        labels[i] = 1 if mask.sum() > 0 else 0
    return labels

print("Building 1-hour labels...")
grid["label_1hr"]  = label_future_window(grid["time"].values, 60)

print("Building 24-hour labels...")
grid["label_24hr"] = label_future_window(grid["time"].values, 24*60)

print("Label distribution (1hr):\n", grid["label_1hr"].value_counts())
print("Label distribution (24hr):\n", grid["label_24hr"].value_counts())


# ── CELL 8: Train/test split (time-based, NOT random) ─────────
# IMPORTANT: for forecasting, split by TIME, not randomly.
# Train on the earlier 80% of the timeline, test on the most recent 20%.
# Random splitting would leak future information into training.

split_idx = int(len(grid) * 0.8)
train_df = grid.iloc[:split_idx]
test_df  = grid.iloc[split_idx:]

print(f"Train: {train_df.shape[0]} steps ({train_df['time'].min()} to {train_df['time'].max()})")
print(f"Test : {test_df.shape[0]} steps ({test_df['time'].min()} to {test_df['time'].max()})")

X_train = train_df[feature_cols]
X_test  = test_df[feature_cols]


# ── CELL 9: Helper to train + evaluate XGBoost for one horizon ─

def train_xgb_for_horizon(label_col, horizon_name):
    y_train = train_df[label_col]
    y_test  = test_df[label_col]

    # handle class imbalance (flares are rarer than "no flare")
    pos = y_train.sum()
    neg = len(y_train) - pos
    scale_pos_weight = neg / pos if pos > 0 else 1

    model = XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        scale_pos_weight=scale_pos_weight,
        eval_metric="logloss",
        random_state=42,
        verbosity=0
    )
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    f1  = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)

    print(f"\n{'='*55}")
    print(f"  XGBoost — Will a C/M/X flare occur in next {horizon_name}?")
    print(f"{'='*55}")
    print(f"  Accuracy : {acc*100:.2f}%")
    print(f"  F1-Score : {f1*100:.2f}%")
    print(f"  ROC-AUC  : {auc:.3f}")
    print(classification_report(y_test, y_pred, target_names=["No flare", "Flare"]))

    return model, y_test, y_pred, y_proba, {"Accuracy": acc*100, "F1": f1*100, "AUC": auc}


# ── CELL 10: Train both horizons ──────────────────────────────
model_1hr, y_test_1hr, pred_1hr, proba_1hr, metrics_1hr = train_xgb_for_horizon("label_1hr", "1 hour")
model_24hr, y_test_24hr, pred_24hr, proba_24hr, metrics_24hr = train_xgb_for_horizon("label_24hr", "24 hours")


# ── CELL 11: Comparison summary ────────────────────────────────
summary = pd.DataFrame({"1 Hour": metrics_1hr, "24 Hours": metrics_24hr}).T
print("\n\nFINAL SUMMARY:")
print(summary.round(2).to_string())


# ── CELL 12: Plots — Confusion matrices + ROC curves ───────────

fig, axes = plt.subplots(2, 2, figsize=(13, 11))

# Confusion matrix - 1hr
cm1 = confusion_matrix(y_test_1hr, pred_1hr)
sns.heatmap(cm1, annot=True, fmt="d", cmap="Blues", ax=axes[0,0],
            xticklabels=["No flare","Flare"], yticklabels=["No flare","Flare"])
axes[0,0].set_title("Confusion Matrix — Next 1 Hour")
axes[0,0].set_xlabel("Predicted"); axes[0,0].set_ylabel("Actual")

# Confusion matrix - 24hr
cm2 = confusion_matrix(y_test_24hr, pred_24hr)
sns.heatmap(cm2, annot=True, fmt="d", cmap="Oranges", ax=axes[0,1],
            xticklabels=["No flare","Flare"], yticklabels=["No flare","Flare"])
axes[0,1].set_title("Confusion Matrix — Next 24 Hours")
axes[0,1].set_xlabel("Predicted"); axes[0,1].set_ylabel("Actual")

# ROC - 1hr
fpr1, tpr1, _ = roc_curve(y_test_1hr, proba_1hr)
axes[1,0].plot(fpr1, tpr1, label=f"AUC = {metrics_1hr['AUC']:.3f}", color="steelblue")
axes[1,0].plot([0,1],[0,1],"--", color="gray")
axes[1,0].set_title("ROC Curve — Next 1 Hour")
axes[1,0].set_xlabel("False Positive Rate"); axes[1,0].set_ylabel("True Positive Rate")
axes[1,0].legend()

# ROC - 24hr
fpr2, tpr2, _ = roc_curve(y_test_24hr, proba_24hr)
axes[1,1].plot(fpr2, tpr2, label=f"AUC = {metrics_24hr['AUC']:.3f}", color="darkorange")
axes[1,1].plot([0,1],[0,1],"--", color="gray")
axes[1,1].set_title("ROC Curve — Next 24 Hours")
axes[1,1].set_xlabel("False Positive Rate"); axes[1,1].set_ylabel("True Positive Rate")
axes[1,1].legend()

plt.tight_layout()
plt.savefig("xgboost_forecast_results.png", dpi=150, bbox_inches="tight")
plt.show()
print("\nSaved plot: xgboost_forecast_results.png ✅")


# ── CELL 13: Feature importance ────────────────────────────────
fig2, axes2 = plt.subplots(1, 2, figsize=(14, 5))

imp1 = pd.Series(model_1hr.feature_importances_, index=feature_cols).sort_values()
imp1.plot(kind="barh", ax=axes2[0], color="steelblue")
axes2[0].set_title("Feature Importance — 1 Hour Model")

imp2 = pd.Series(model_24hr.feature_importances_, index=feature_cols).sort_values()
imp2.plot(kind="barh", ax=axes2[1], color="darkorange")
axes2[1].set_title("Feature Importance — 24 Hour Model")

plt.tight_layout()
plt.savefig("xgboost_feature_importance.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved plot: xgboost_feature_importance.png ✅")
```
